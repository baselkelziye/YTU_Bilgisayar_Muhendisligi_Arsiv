## Exercise 1 
Create a 2D Gaussian kernel using the formula in the slides (set sigma^2=2, the center is in the middle of the kernel. Make your kernel size odd).

Formula:

$$G\sigma(x,y)=(1/(2*\pi*\sigma^2)) * \exp(-(x^2 + y^2)/(2 \sigma^2))$$

```python
def calculateKernelSize(sigma):
    kernel_size = int(6 * sigma)
    
    # Çiftse tek yap
    if ((kernel_size % 2) == 0):
        kernel_size += 1

    print("kernel_size: ", kernel_size)
    return kernel_size

def twoDimensionalGaussianFilter(sigma, kernel_size):
    number_of_samples = kernel_size
    max_range = int(number_of_samples / 2)
    range_vector = numpy.linspace(-max_range, max_range, number_of_samples)
    x , y = numpy.meshgrid(range_vector, range_vector)
    # print("x:")
    # print(x)
    # print("y:")
    # print(y)
    kernel = (1. / (2. * numpy.pi * sigma**2)) * numpy.exp(-(x**2 + y**2)/(2 * sigma**2))
    return kernel


# TODO: Change kernel size

sigma = numpy.sqrt(2)
calculateKernelSize(sigma)
kernel_size = 11
print("sigma:", sigma)
print (twoDimensionalGaussianFilter(sigma, kernel_size))
```
