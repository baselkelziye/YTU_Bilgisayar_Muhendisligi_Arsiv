Create 1D Gauss Kernel
----------------------

$$g(x)=a \exp(-x^2/(2 \sigma^2))$$

```python
def gauss1(sigma, width):
    
    hwidth = round((width-1)/2)
    x = np.arange(-hwidth, hwidth+1,1)
    
    g = np.exp(-x**2/(2*sigma**2))
    return g/np.sum(g)
    
    
g1=gauss1(2,11)
print(g1)
g1=np.reshape(g1,(1,-1))
print(g1)
    
```

```python
plt.imshow(g1)
```

```python
f1 = cv2.filter2D(f,-1,g1)
```
